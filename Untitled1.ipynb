```python

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Imports
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image, ImageChops, ImageEnhance

from keras.models import Sequential, load_model
from keras.layers import Conv2D, MaxPool2D, Dropout, Flatten, Dense
from keras.utils.np_utils import to_categorical
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint, EarlyStopping

# Set base path to your CASIA2.0 dataset in Google Drive
base_path = '/content/drive/MyDrive/CASIA2'

# ELA image creation
def convert_to_ela_image(path, quality=98):
    temp_filename = 'temp_file.jpg'
    image = Image.open(path).convert('RGB')
    image.save(temp_filename, 'JPEG', quality=quality)
    temp_image = Image.open(temp_filename)
    ela_image = ImageChops.difference(image, temp_image)
    extrema = ela_image.getextrema()
    max_diff = max([ex[1] for ex in extrema]) or 1
    scale = 255.0 / max_diff
    ela_image = ImageEnhance.Brightness(ela_image).enhance(scale)
    return ela_image

# Image preprocessing
def prepare_image(image_path):
    return np.array(convert_to_ela_image(image_path).resize((128, 128))).flatten() / 255.0

# Load dataset
def load_dataset(folder_path, label, exts=('jpg', 'bmp', 'tif')):
    data, labels = [], []
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith(exts):
                full_path = os.path.join(root, file)
                data.append(prepare_image(full_path))
                labels.append(label)
    return data, labels

# Load training and test data
X_train_real, Y_train_real = load_dataset(os.path.join(base_path, 'Au_Train'), 0)
X_train_fake, Y_train_fake = load_dataset(os.path.join(base_path, 'Tp_Train'), 1)
X_test_real, Y_test_real = load_dataset(os.path.join(base_path, 'Au_Test'), 0)
X_test_fake, Y_test_fake = load_dataset(os.path.join(base_path, 'Tp_Test'), 1)

# Combine datasets
X_train = np.array(X_train_real + X_train_fake)
Y_train = to_categorical(Y_train_real + Y_train_fake, 2)
X_test = np.array(X_test_real + X_test_fake)
Y_test = to_categorical(Y_test_real + Y_test_fake, 2)

# Reshape for CNN input
X_train = X_train.reshape(-1, 128, 128, 3)
X_test = X_test.reshape(-1, 128, 128, 3)

# Define CNN model
def build_model():
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
        Conv2D(32, (3, 3), activation='relu'),
        Conv2D(32, (7, 7), activation='relu'),
        MaxPool2D(pool_size=(2, 2)),
        Dropout(0.25),
        Flatten(),
        Dense(256, activation='relu'),
        Dropout(0.5),
        Dense(2, activation='softmax')
    ])
    return model

# Compile model
model = build_model()
model.compile(optimizer=Adam(learning_rate=1e-5), loss='binary_crossentropy', metrics=['accuracy'])

# Define callbacks
checkpoint = ModelCheckpoint('/content/drive/MyDrive/model_best_casia98.h5', monitor='val_accuracy', save_best_only=True)
early_stop = EarlyStopping(monitor='val_accuracy', patience=20)

# Train the model
history = model.fit(X_train, Y_train, batch_size=64, epochs=100, validation_data=(X_test, Y_test),
                    callbacks=[checkpoint, early_stop])

# Save final model
model.save('/content/drive/MyDrive/model_final_casia98.h5')
```
